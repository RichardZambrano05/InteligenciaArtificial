```python
import pandas as pd

# Cargar el archivo CSV
file_path = "ruta_a_tu_archivo.csv"
df = pd.read_csv(file_path)

# Ver las primeras filas del DataFrame para entender la estructura
print(df.head())

# Nombre de la columna que quieres evaluar
columna_original = 'nombre_de_la_columna'

# Crear una nueva columna basada en la condición
df['nueva_columna'] = df[columna_original].apply(lambda x: 1 if x > 36 else 0)

# Guardar el DataFrame modificado en un nuevo archivo CSV
df.to_csv('ruta_a_tu_archivo_modificado.csv', index=False)

# Ver las primeras filas del DataFrame modificado
print(df.head())
```

```python
#En base al analisis anterior vamos a realizar otro tipo de grafico para tener
#mas claro y ya sabemos que tenemos que enfocarnos en los petalos
setosa=df[df["species"]=='Iris-setosa']
versicolor=df[df['species']=='Iris-versicolor']
virginica=df[df['species']=='Iris-virginica']
#y vamos a graficar ahora los petalos
```

```python
setosa.shape
```

```python
#Se definio el area de trabajo para la figura
import matplotlib.pyplot as plt
figure,axis=plt.subplots()
figure.set_size_inches(10,7)

axis.set_title("Petal Length vs Petal Width")
axis.set_xlabel("Petal Length")
axis.set_ylabel("Petal Width")

#usamos axis para dar etiquetas a los ejes
axis.scatter(setosa["petal_length"],setosa["petal_width"],color="green",label="Setosa")
axis.scatter(versicolor["petal_length"],versicolor["petal_width"],color="red",label="Versicolor")
axis.scatter(virginica["petal_length"],virginica["petal_width"],color="blue",label="Virginica")

axis.legend()

#con esto queda claro que vamos a ocupar solo dos variables que son acncho y
#longitud de los petalos
```

```python
#vamos a darle a las cadenas de caracteres numeros para poder realizar el
#entrenamiento donde dice especie
df.head(3)
```

```python
#Creamos una varibale temporal para cambiar a valores numericos
#y en target se añadira lo que hay en la variable temporal
tmp=list()
for species in df["species"]:
  if species=="Iris-setosa":
    tmp.append(1)
  elif species=="Iris-versicolor":
    tmp.append(2)
  else:
    tmp.append(3)
df["target"]=tmp
df.head(3)
```

```python
#Verificamos que se realizo el cambio
df['target'].unique()
```

```python
#creamos variables X y Y para los petalos
#no incluyo estas varibales
x=df.drop(["sepal_length","sepal_width","species",'target'],axis=1)
x.head(3)
```

```python
#Apunto a target Y
y=df["target"]
y.head(3)
```

```python
#Ahora vamos hacer uso de los algoritmos de IA Machine Learning, donde vamos a
#dividir el 70 porc para el aprendijaze y el 30 para evaluacion(0.3)
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
#ya esta dividivo
```

```python
#Empezamos con el entrenamiento del modelo
model.fit(x_train,y_train)
```

```python
#utilizamos el moedelo con el 30 pocr
yPredictions=model.predict(x_test)
```

```python
yPredictions
```

```python
#EVALUACION
from sklearn.metrics import accuracy_score
accuracy_score(y_test,yPredictions)
```

```python
#El modelo ya esta entrenado, ahora vamos a guardarlo con pickle
import pickle
pickle.dump(model,open("modelo.pkl","wb"))
```
